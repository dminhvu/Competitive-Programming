```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('melbourne_housing_full.csv')
df
```

```python
df.columns
```

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Distance','Bedroom2','Bathroom','Car',
               'Landsize','BuildingArea','CouncilArea','Regionname','Propertycount','Price']
```

```python
df = df[cols_to_use]
df
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero = ['Bedroom2','Car','Propertycount','Bathroom','Distance']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
df
```

```python
df.isna().sum()
```

```python
cols_to_fill_mean = ['Landsize','BuildingArea']
for col in cols_to_fill_mean:
    df[col] = df[col].fillna(df[col].mean())
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df = pd.get_dummies(df,drop_first=True)
df
```

```python
X = df.drop(['Price'],axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
reg.score(X_train,y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=1,max_iter=100,tol=0.1)
lasso_reg.fit(X_train,y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=1,max_iter=500,tol=0.001)
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.score(X_test,y_test)
```

```python
ridge_reg.score(X_train,y_train)
```
